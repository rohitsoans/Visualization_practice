```python
plt.subplots(figsize=(10,10))
sns.set_style("darkgrid")
hw=sns.boxplot(data=mov3, x="Genre",y="Gross%US",orient='v',color="lightgray")
bw=sns.stripplot(data=mov3, x="Genre",y="Gross%US", jitter=True, hue="Studio")
plt.title("Domestic Gross% by Genre",size=20)
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Gross % US",fontsize=20)
plt.legend(loc="upper right")
```
